```python
import pandas as pd
```

```python
df = pd.read_csv("오치동기온.csv")
df
```

```python
df.columns = ['day', 'hour', '기온']
```

```python
df
```

```python
df.groupby('day')['기온'].max()
```

```python
df.groupby('day')['기온'].min()
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'batang'
matplotlib.rcParams['font.size'] =  15
matplotlib.rcParams['axes.unicode_minus'] = True
```

```python
max_temp = df.groupby('day')['기온'].max()
min_temp = df.groupby('day')['기온'].min()

# 그래프 그리기
plt.figure(figsize=(8,5))
max_temp.plot( color='red', marker='o')  # 막대 그래프
for x, y in zip(max_temp.index, max_temp.values):
    plt.text(x, y + 0.2, f'{y:.1f}℃', ha='center', va='bottom', fontsize=10)
min_temp.plot( color='blue', marker='o')
for x, y in zip(min_temp.index, min_temp.values):
    plt.text(x, y - 0.5, f'{y:.1f}℃', ha='center', va='top', fontsize=10)

plt.title('오치동 1월 1주 날짜별 최고 기온')

plt.xlabel('1월 1주')
plt.ylabel('기온 (℃)')
plt.show()
```
