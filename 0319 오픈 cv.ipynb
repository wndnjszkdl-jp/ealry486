1. 이미지 읽기 옵션
    - cv.2 IMREAD_COLOR : 컬러 이미지, 투명 영역은 무시
    - cv.2 IMREAD_GRAYSCALE : 흑백이미지
    - cv.2 IMREDA_UNCHANGED : 투명영역까지 포함

```python
import cv2 
img_color = cv2.imread('image.jpg', cv2.IMREAD_COLOR)
img_gray = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
img_unchanged = cv2.imread('image.jpg', cv2.IMREAD_UNCHANGED)
```

```python
cv2.imshow('img_color', img_color)
cv2.imshow('img_gray', img_gray)
cv2.imshow('img_unchanged', img_unchanged)

if cv2.waitKey(0) == ord('A'):
    print('사용자가 A를 눌렀음.')

cv2.destroyAllWindows
```

4. 직선그리기
    - cv2.LINE_4 : 상하좌우 4가지 방향으로만 연결된 선
    - cv2.LINE_8 : 상하좌우 + 대각선 포함 8가지 방향으로 연결된 선(DEFALT)
    - cv2.LINE_AA : 부드러운 선(anti-aliasing)


- cv2.line(그릴위치 , 시작점, 끝점, 색깔, 두께, 선 종류)

```python
import cv2
import numpy as np

# 세로 480 * 가로 640 , 3 채널 [RGB]에 해당하는 빈 스케치북 만들기

img = np.zeros((480,640,3), dtype=np.uint8)
img[:] = 255
COLOR = (0, 0, 255)
THIC = 3

cv2.line( img, (50,400), (400,200), COLOR, THIC, cv2.LINE_AA)
cv2.line( img, (50,200), (400,400), COLOR, THIC, cv2.LINE_AA)

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows
```

5. 원그리기
    - cv2.circle(원의 중심점, 반지름값, 색깔, 두께, 선 종류)



```python
import cv2
import numpy as np

# 세로 480 * 가로 640 , 3 채널 [RGB]에 해당하는 빈 스케치북 만들기

img = np.zeros((480,640,3), dtype=np.uint8)
img[:] = 255
COLOR = (0, 180, 30)
THIC = 10
RADIUS = 50

cv2.circle(img, (300,200), RADIUS, COLOR, THIC, cv2.LINE_AA)
cv2.circle(img, (300,200), RADIUS, COLOR, cv2.FILLED, cv2.LINE_AA)

cv2.imshow('circle',img)
cv2.waitKey(0)
cv2.destroyAllWindows
```

 6. 사각형 그리기
     - cv2.rectangle( 그릴위치, 왼쪽위 좌표, 오른쪽 아래 좌표, 색깔, 두께 )

```python
import cv2
import numpy as np

# 세로 480 * 가로 640 , 3 채널 [RGB]에 해당하는 빈 스케치북 만들기

img = np.zeros((480,640,3), dtype=np.uint8)
img[:] = 255
COLOR = (0, 180, 30)
THIC = 10

cv2.rectangle( img, (50,300), (300,50), COLOR, THIC, cv2.LINE_AA)
cv2.rectangle( img, (100,250), (250,100), COLOR, THIC, cv2.LINE_AA)
cv2.rectangle( img, (150,200), (200,150), COLOR, THIC, cv2.LINE_AA)

cv2.imshow('circle',img)
cv2.waitKey(0)
cv2.destroyAllWindows
```

7. 다각형 그리기
    - cv2.polylines( 위치, 그릴 좌표들 , 닫힘여부(T/F), 색깔, 두께, 선종류)
    - cv2.fillpoly( 위치, 그릴 좌표들 , 색깔, 선 종류)

```python
import cv2
import numpy as np

# 세로 480 * 가로 640 , 3 채널 [RGB]에 해당하는 빈 스케치북 만들기

img = np.zeros((480,640,3), dtype=np.uint8)
COLOR = (0, 255, 80)
THIC = 10
pt1 = np.array([[[100,100], [200,100], [100,200]],[[200,100], [300,100], [300,200]]])
pt2 = np.array([[[100,300], [200,300], [100,400]],[[200,300], [300,300], [300,400]]])
pt3 = np.array([[450, 150],[550, 150],[600, 240],[550, 330],[450, 330],[400, 240]], dtype=np.int32)

# pt2 = np.array([[200,100], [300,100], [300,200]])

cv2.polylines( img, pt1 , True, COLOR, THIC, cv2.LINE_AA)
# cv2.polylines( img, [pt2] , True, COLOR, THIC, cv2.LINE_AA)

cv2.fillPoly( img, pt2 , COLOR,  cv2.LINE_AA)
# cv2.fillPoly( img, pt1 , COLOR,  cv2.LINE_AA)
cv2.fillPoly( img, [pt3] , COLOR,  cv2.LINE_AA)


cv2.imshow('polylines',img)
cv2.waitKey(0)
cv2.destroyAllWindows
```
