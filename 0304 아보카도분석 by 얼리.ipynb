```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
df = pd.read_csv('avocado.csv')
df
```

```python
유기농과일반 = df.groupby('type')['AveragePrice'].mean()
```

```python
지역별평균가격 = df.groupby('region')['AveragePrice'].mean()
지역별평균가격
```

```python
지역별평균가격상위10 = 지역별평균가격.sort_values(ascending=False).head(10)
지역별평균가격상위10
```

```python
지역별평균가격상위10.plot(kind='bar', color='skyblue')

plt.title("지역별 평균 아보카도 가격")
plt.xlabel("지역")
plt.ylabel("평균 가격 (USD)")
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
유기농과일반.plot(kind='bar', color='saddlebrown')

plt.title("유기농과 일반의 평균 아보카도 가격 차이")
plt.xlabel("유기농 = organic")
plt.ylabel("평균 가격 (USD)")
plt.grid(axis='y', alpha=0.3)
plt.xticks(rotation=360)
plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
# 2. 2016년 데이터만 필터링
df_2016 = df[(df['Date'].dt.year == 2016)]

df_2016
```

```python
df_2016['Month'] = df_2016['Date'].dt.month
```

```python
df_2016
```

```python
월별판매량평균 = df_2016.groupby('Month')['Total Volume'].mean()
월별판매량평균
```

```python
plt.figure(figsize=(10,5))
월별판매량평균.plot(marker='o', color='#A2C523')  # 아보카도 색
plt.title("2016년 월별 평균 아보카도 판매량")
plt.xlabel("월")
plt.ylabel("판매량")
plt.xticks(range(1,13))
plt.grid(alpha=0.3)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
월별아보카도가격평균 = df_2016.groupby('Month')['AveragePrice'].mean()
월별아보카도가격평균
```

```python
plt.figure(figsize=(10,5))
월별아보카도가격평균.plot(marker='o', color='#A2C523')  # 아보카도 색
plt.title("2016년 월별 평균 아보카도 평균 가격")
plt.xlabel("월")
plt.ylabel("가격")
plt.xticks(range(1,13))
plt.grid(alpha=0.3)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
월별사이즈별평균 = df_2016.groupby('Month')[['4046','4225','4770']].mean()
월별사이즈별평균
```

```python
plt.figure(figsize=(12,6))

# 3가지 사이즈 선 그리기
plt.plot(월별사이즈별평균.index, 월별사이즈별평균['4046'], marker='o', label='Small (4046)', color='lightgreen')
plt.plot(월별사이즈별평균.index, 월별사이즈별평균['4225'], marker='o', label='Medium (4225)', color='orange')
plt.plot(월별사이즈별평균.index, 월별사이즈별평균['4770'], marker='o', label='Large (4770)', color='brown')

plt.title("2016년 월별 아보카도 사이즈별 평균 판매량")
plt.xlabel("월")
plt.ylabel("판매량")
plt.xticks(range(1,13))
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

```python
년도별사이즈별평균 = df.groupby('year')[['4046','4225','4770']].mean()
년도별사이즈별평균
```

```python
plt.figure(figsize=(10,6))

# 3가지 사이즈 선 그리기
plt.plot(년도별사이즈별평균.index, 년도별사이즈별평균['4046'], marker='o', label='Small (4046)', color='lightgreen')
plt.plot(년도별사이즈별평균.index, 년도별사이즈별평균['4225'], marker='o', label='Medium (4225)', color='orange')
plt.plot(년도별사이즈별평균.index, 년도별사이즈별평균['4770'], marker='o', label='Large (4770)', color='brown')

plt.title("2015~2018년 년도별 아보카도 사이즈별 평균 판매량")
plt.xlabel("년")
plt.ylabel("판매량")
plt.xticks(년도별사이즈별평균.index)
plt.grid(alpha=0.2)
plt.legend()
plt.show()
```

```python
월별_판매량 = pd.pivot_table(
    df_2016,
    index='Month',            # 행: 월
    columns='type',           # 열: conventional / organic
    values='Total Volume',    # 계산할 값: 판매 무게
    aggfunc='mean',            # 합계             # 결측치 0으로
)
```

```python
plt.figure(figsize=(10,6))
plt.plot(월별_판매량.index, 월별_판매량['conventional'], marker='o', label='일반', color='skyblue')
plt.plot(월별_판매량.index, 월별_판매량['organic'], marker='o', label='유기농', color='green')


plt.title("2016년 월별 일반 vs 유기농 아보카도 판매량 추이")
plt.xlabel("월")
plt.ylabel("총 판매량")
plt.xticks(월별_판매량.index)
plt.ticklabel_format(style='plain', axis='y')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
