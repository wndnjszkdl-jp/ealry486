```python
# 버전 1
import cv2
cap = cv2.VideoCapture('video2.mp4')
codec = cv2.VideoWriter_fourcc(*'DIVX')

ret, frame = cap.read()
frame = cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE)
height, width = frame.shape[:2]

movie = cv2.VideoWriter('video_out.avi', codec, 60, (width,height))

while True:
    if not ret :
        print('더 이상 가져올 이미지가 없다.')
        break
        
    movie.write(frame)
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(1) == ord('q'):
        print('사용자에 의해 종료됨.')
        break
    ret, frame = cap.read()
    frame = cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE)

        
cap.release()
movie.release()
cv2.destroyAllWindows()
```

```python
# 버전 2
import cv2
cap = cv2.VideoCapture('video2.mp4')
fps = cap.get(cv2.CAP_PROP_FPS) #약 30
delay = round(1000 / (fps*2))
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) #540
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) #960
fourcc = cv2.VideoWriter_fourcc(*'DIVX') #코덱 설정
out = cv2.VideoWriter('video_out2.avi', fourcc, fps*3, (height,width))

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret :
        break
        
    frame2 = cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE)
    
    cv2.imshow('daramgwe', frame2)
    out.write(frame2)
    
    if cv2.waitKey(1) == ord('q') :
        print('사용자에 의해 종료되었습니다.')
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
