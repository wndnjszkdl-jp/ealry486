```python
import cv2
import numpy as np
img = cv2.imread('card.jpg')
point4 = []
COLOR = (0,0,255)

def show_result():
    print("점 4개 선택 완료:", point4)
    
def mouse_handler(event, x, y, flag, param ):
    if event == cv2.EVENT_LBUTTONDOWN:
        point4.append((x,y))
            
        if len(point4) == 4:
            show_result()

        for point in point4:
            cv2.circle(img, point, 10, COLOR, cv2.FILLED)
            cv2.imshow('img', img)

cv2.namedWindow('img')
cv2.setMouseCallback('img', mouse_handler)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import math
img = cv2.imread('nicoma.jpg')
point4 = []
COLOR = (0,0,255)

    
def mouse_handler(event, x, y, flag, param ):
    if event == cv2.EVENT_LBUTTONDOWN:
        point4.append((x,y))
            
        if len(point4) == 4:
            show_result()

        for point in point4:
            cv2.circle(img, point, 1, COLOR, cv2.FILLED)
            cv2.imshow('img', img)

        cv2.imshow('img', img)


def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def show_result():
    dists = []
    for i in range(4):
        d = distance(point4[i], point4[(i+1) % 4])
        dists.append(d)
    
    # 가로, 세로 구하기
    height = int(max(dists))
    width = int(min(dists))
    src = np.float32(point4)
    dst = np.array([[0,0], [width,0], [width, height], [0,height]], dtype=np.float32)
    
    
    matrix = cv2.getPerspectiveTransform(src, dst)
    res = cv2.warpPerspective(img, matrix, (width, height) )
    cv2.imshow('result', res) 

cv2.namedWindow('img')
cv2.setMouseCallback('img', mouse_handler)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
