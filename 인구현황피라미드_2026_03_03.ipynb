##### 판다스, 맵플롯립, 맵플롯립 한글 설정

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] =  15
matplotlib.rcParams['axes.unicode_minus'] = False
```

- 파일 이름 설정

```python
filename = '2016.xlsx'
```

- 남자 데이터 파일 읽어오기(skiprows, usecols, index_col)

```python
df_m = pd.read_excel( filename, skiprows=3, usecols='b, e:y',index_col='행정기관' )
df_m.head()
```

- 전국 인구 (콤마), 제외하고 숫자로 변환

```python
df_m.info()
```

```python
df_m.iloc[0]
```

```python
cols = df_m.loc[:, '0~4세':'100세 이상'].columns
```

```python
df_m[cols] = df_m[cols].replace(',', '',regex=True).astype(int)
df_m
```

```python
df_m.info()
```

```python
df_m.iloc[0]
```

```python
df_m.iloc[0] = df_m.iloc[0]
df_m
```

- 여자 데이터 파일 읽어오기

```python
df_w = pd.read_excel( filename, skiprows=3, usecols='b, ab:av',index_col='행정기관' )
df_w.head(3)
```

- 남자 데이터와 여자 데이터 컬럼명 일치시키기

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.head(3)
```

- 전국 인구 (콤마), 제외하고 숫자로 변환

```python
cols2 = df_w.loc[:, '0~4세':'100세 이상'].columns
```

```python
df_w[cols2] = df_w[cols2].replace(',', '',regex=True).astype(int)
df_w
```

```python
df_m.info()
```

- 인구 피라미드 시각화, 크기( 15, 9 ), barh, 제목(2016년 인구 피라미드), png 파일 저장

```python
plt.figure(figsize=(15,9))
m_values = -df_m.iloc[0] // 1000   # 남자 (음수, 단위: 천 명 → 만 명 아님)
w_values = df_w.iloc[0] // 1000
plt.barh(df_m.columns, m_values,  label='남자')
plt.barh(df_w.columns, w_values, label='여자')
plt.title('2016년 인구 피라미드')
plt.legend() 
for i, v in enumerate(m_values):
    plt.text(v - 3, i, f"{abs(v)//10}만", va='center', ha='right', fontsize=11)

for i, v in enumerate(w_values):
    plt.text(v + 3, i, f"{v//10}만", va='center', ha='left', fontsize=11)

plt.savefig('2016년 인구 피라미드.png', dpi=100)
plt.show()
```
