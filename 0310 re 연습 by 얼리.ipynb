```python
text = "내 번호는 010-1234-5678이야. 전화해!"
```

- text에서 핸드폰 번호 가운데 숫자(1234)만 **** 문자로 마스킹

```python
text = """
홍길동 생일: 1990/05/21
김철수 생일: 1985-11-03
"""
```

- text에서 날짜 데이터를 추출해서 1990년 05월 21일 형식으로 통일해서 출력

```python
p = re.sub(r'(\d{4}).(\d{2}).(\d{2})', r'\1년\2월\3일', text)
print(p)
```

```python
text = "그가 '파이썬'은 정말 '재밌다'고 말했다."
```

- text에서 작은따옴표(') 안에 있는 글자만 추출 (결과 : ['파이썬', '재밌다'] )

```python
m = re.findall(r"'(.*?)'", text)
m
```

```python
html_text = """
<div class="content">
    <h1>파이썬 정규표현식</h1>
    <p>이 글은 <b>정말</b> 유익하네요! <a href="link">링크</a></p>
</div>
"""
```

- html_text에서 태그(<  >)를 모두 삭제
- 결과 => 파이썬 정규표현식
-         이 글은 정말 유익하네요! 링크

```python
p = re.sub(r'<.*?>', " ", html_text)

print(p)
```

```python
text = "사과 1500원, 바나나 3달러, 포도 2000원, 배 5유로"
```

- text에서 원화 가격(1500, 2000)만 추출 ( 전방 탐색 )

```python
m = re.findall(r'\d+(?=원)', text)
m
```

```python
log_data = """
User logged in from IP: 192.168.1.10 at 10:00 AM.
Failed login attempt from 10.0.0.255!
Warning: device 1234.5678.90 is not an IP.
"""
```

- log_data에서 IP 주소만 추출
- 결과 : ['192.168.1.10', '10.0.0.255']

```python
m = re.findall(r'\d+\.\d+\.\d+\.\d+', log_data)
m
```

```python
text = "I love love Python programming. It is is very fun!"
```

- 연속적인 중복 단어 제거 ( 후방 탐색 )
