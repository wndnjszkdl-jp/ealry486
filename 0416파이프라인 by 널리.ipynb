```python
import tf_keras as keras
from tf_keras import layers
import tensorflow_hub as hub
```

```python
hub_ef = keras.Sequential([
    layers.Input(shape=(224,224,3)),
    layers.Rescaling(1.0/255.0),
    hub.KerasLayer("https://www.kaggle.com/models/tensorflow/efficientnet/TensorFlow2/b0-classification/1")
])
```

```python
import numpy as np
from PIL import Image

dog_png = np.array(Image.open('images/dog.png'))
```

```python
from keras.applications import efficientnet

predict = hub_ef.predict(dog_png[np.newaxis, : ])
efficientnet.decode_predictions(predict)
```
