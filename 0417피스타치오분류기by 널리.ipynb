```python
import numpy as np
import tensorflow_hub as hub
from PIL import Image
import tf_keras as keras
from tf_keras import layers
from keras.applications import efficientnet
```

```python
pistachio_sample = np.array(Image.open('Pistachio_Image_Dataset/Kirmizi_Pistachio/kirmizi (1).jpg'))
pistachio_sample.shape
```

```python
train_ds, val_ds = keras.utils.image_dataset_from_directory('Pistachio_Image_Dataset',image_size=(224,224), batch_size=16, validation_split=0.2, subset='both', seed=42)
```

```python
hub_ef = keras.Sequential([
    layers.Input(shape=(224,224,3)),
    layers.Rescaling(1.0/255.0),
    hub.KerasLayer("https://www.kaggle.com/models/tensorflow/efficientnet/TensorFlow2/b0-feature-vector/1",
    trainable=False)
])
```

```python
model = keras.Sequential([
    hub_ef,
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
rmsprop = keras.optimizers.RMSprop ( learning_rate=5e-5)

model.compile( optimizer = rmsprop, loss='binary_crossentropy', metrics=['accuracy'])

hist = model.fit ( train_ds , epochs=20, validation_data=val_ds)
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt

plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.legend(['train', 'val'])
plt.title('Loss')
plt.show()
```

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.legend(['train', 'val'])
plt.title('Accuracy')
plt.show()
```
