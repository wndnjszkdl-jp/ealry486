```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
# 날짜
dates = pd.to_datetime([
    "2023-01-01","2023-07-01","2024-01-01",
    "2024-07-01","2025-01-01","2025-07-01",
    "2026-01-01"
])

# 기사 기반 대표 메뉴 가격 데이터
bbq      = [20000, 20000, 20500, 23000, 23000, 24000, 25000]
bhc      = [17000, 18000, 19000, 20000, 20000, 21000, 23000]
kyochon  = [19000, 19000, 19000, 20000, 22000, 24000, 26000]
pelicana = [18000, 18000, 19000, 19000, 19000, 20000, 20000]  # 추정

df = pd.DataFrame({
    "B0Q": bbq,
    "B0C": bhc,
    "교0": kyochon,
    "페0카나": pelicana
}, index=dates)

plt.figure(figsize=(10,6))
for col in df.columns:
    plt.plot(df.index, df[col], marker='o', label=col)
    
    for x, y in zip(df.index, df[col]):
        plt.text(x, y + 150,          # y+150 → 숫자가 점 위에 떠 있도록
                 f"{y:,}원",
                 ha='center', va='bottom',
                 fontsize=9)


plt.title("대표 메뉴 기준 4대 치킨 가격 비교(예시)", fontsize=14)
plt.xlabel("기간", fontsize=12)
plt.ylabel("가격 (원)", fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
