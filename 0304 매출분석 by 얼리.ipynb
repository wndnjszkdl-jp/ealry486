- 라이브러리 임포트

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
np.random.seed(0)
```

- 가상의 상품 정보 데이터 만들기 (Products)
- 

```python
Product = pd.DataFrame({
    'ProductID': ['P01', 'P02', 'P03'],
    'Category' : ['Electronics', 'Clothing', 'Home'],
    'Price': [50000, 30000, 15000]
})

Product
```

- 100일 치의 가상 주문 내역 데이터 만들기 (Orders)

```python
Date = pd.date_range('2023-01-01', freq='d', periods=100 )
Date
```

- 상품정보와 주문내역 정보 출력
-  

```python
Quantity = np.random.randint(1, 6, size=100)  # 1~5, 100개
Quantity
```

```python
product_list = ['P01', 'P02', 'P03']
ProductID = np.random.choice(product_list, size=100, replace=True)
ProductID
```

```python
Orders = pd.DataFrame({
    'Date': Date,
    'ProductID' : ProductID,
    'Quantity': Quantity
})

Orders
```

```python
df = pd.merge(Orders, Product, on='ProductID', how='right').sort_values(by='Date')


df.head(10)
```

- 상품정보와 주문내역 합치기(merge)
- orders를 기준으로 products 데이터를 가져와서 붙임

```python
df = pd.merge(Orders, Product, on='ProductID', how='right').sort_values(by='Date')


df
```

- 매출액(Revenue) 계산하기 (가격 x 수량)

```python

df['Revenue'] = df['Price'] * df['Quantity']

df
```

- 월(Month) 정보만 따로 뽑아내기

```python
df['Month'] = df['Date'].dt.month
df
```

- 합쳐진 데이터 출력

```python
df
```

- 피벗 테이블 만들기( 행 : 월, 열 : 카테고리, 값 : 매출액, 계산함수 : 합계 )

```python
카테고리별매출액 = pd.pivot_table(
    df,
    index='Month',            # 행: 월
    columns='Category',           # 열: conventional / organic
    values='Revenue',    # 계산할 값: 판매 무게
    aggfunc='sum',            # 합계             # 결측치 0으로
)

카테고리별매출액
```

- 데이터가 비어있는 곳(NaN)은 0으로 채우기

- 피벗테이블 출력

```python
카테고리별매출액.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set1')
plt.title('카테고리별 매출액')
plt.xticks(rotation=0)  # 0 = 가로, 90 = 세로, 45 = 대각선 등
plt.ticklabel_format(style='plain', axis='y')
```
