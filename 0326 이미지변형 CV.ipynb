- 임계값(Threshold)

```python
img = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)

def evt(val):
    print(val)
    pass
    
name = 'Trackbar'
cv2.namedWindow(name)

cv2.createTrackbar('threshold', name, 127, 255, evt)

while True :
    thresh = cv2.getTrackbarPos('threshold', name )
    ret, binary = cv2.threshold(img, thresh, 255, cv2.THRESH_BINARY)
    #return, data
   
    if not ret :
        break
    
    cv2.imshow(name, binary)
    if cv2.waitKey(1) == ord('q'):
        break
    
cv2.destroyAllWindows()
```

 - Adaptive threshold

```python
import cv2
img = cv2.imread('book2.jpg', cv2.IMREAD_GRAYSCALE)

def evt(val):
    print(val)
    pass
    
name = 'Trackbar'
cv2.namedWindow(name)
cv2.createTrackbar('block_size', name, 25, 100, evt)
# block_size의 값은 1보다 큰 홀수
cv2.createTrackbar('c', name, 3, 10, evt)

while True :
    block_size = cv2.getTrackbarPos('block_size', name )
    c = cv2.getTrackbarPos('c', name )

    if block_size <= 1 :
        block_size = 3

    if block_size % 2 ==0:
        block_size += 1
        
    binary = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, block_size, c)
    #return, data
   
    
    cv2.imshow(name, binary)
    if cv2.waitKey(1) == ord('q'):
        break
    
cv2.destroyAllWindows()
```

 - 오츠(otsu) : 최적의 임계값을 찾아주는 알고리즘

```python
import cv2

img = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)

ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
    #return, data
ret, otsu = cv2.threshold(img, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print('otsu : ', ret)

cv2.imshow('img', img)
cv2.imshow('binary', binary)
cv2.imshow('otsu', otsu)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

### 이미지 노이즈 제거
 - dilate(팽창) : 이미지를 확장해서 작은 노이즈를 제거

```python
import numpy as np

kernel = np.ones((3,3), dtype = np.uint8)
kernel
```

```python
import cv2
img = cv2.imread('dilate.png' , cv2.IMREAD_GRAYSCALE)

dilate1 = cv2.dilate(img, kernel, iterations=1)
dilate2 = cv2.dilate(img, kernel, iterations=2)
dilate3 = cv2.dilate(img, kernel, iterations=3)

cv2.imshow('img', img)
cv2.imshow('dilate1', dilate1)
cv2.imshow('dilate2', dilate2)
cv2.imshow('dilate3', dilate3)

cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
import cv2
img = cv2.imread('erode.png' , cv2.IMREAD_GRAYSCALE)

erode1 = cv2.erode(img, kernel, iterations=1)
erode2 = cv2.erode(img, kernel, iterations=2)
erode3 = cv2.erode(img, kernel, iterations=3)

cv2.imshow('img', img)
cv2.imshow('erod1', erode1)
cv2.imshow('erod2', erode2)
cv2.imshow('erod3', erode3)

cv2.waitKey(0)
cv2.destroyAllWindows
```

 - 이미지 변환 (열림 & 닫힘)
 - 열림 : 침식 후 팽창 , 노이즈 깎아서 제거 후 팽창
 - 열림 dilate(erode(image)) 

```python
import numpy as np

kernel = np.ones((3,3), dtype = np.uint8)
kernel

import cv2
img = cv2.imread('erode.png' , cv2.IMREAD_GRAYSCALE)
erode = cv2.erode(img, kernel, iterations=1)
dilate = cv2.dilate(erode, kernel, iterations=1)

cv2.imshow('img', img)
cv2.imshow('erode', erode)
cv2.imshow('dilate', dilate)

cv2.waitKey(0)
cv2.destroyAllWindows
```

 ### 이미지 검출 - 경계선
 - Canny Edge Detection

```python
img = cv2.imread('img.jpg')

def evt(val):
    print(val)
    pass

name = 'Trackbar'
cv2.namedWindow(name)

cv2.createTrackbar('threshold1', name, 0, 255, evt)
cv2.createTrackbar('threshold2', name, 0, 255, evt)

while True :
    threshold1 = cv2.getTrackbarPos('threshold1', name )
    threshold2 = cv2.getTrackbarPos('threshold2', name )
    
    
    canny = cv2.Canny(img, threshold1, threshold2)

    cv2.imshow(name, canny)
    if cv2.waitKey(1) == ord('q'):
        break
        

cv2.destroyAllWindows()
```

### 이미지 검출 - 윤곽선
- 윤곽선(contour)

```python
import cv2

img = cv2.imread('bcard.jpg')

temp = img.copy()

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, otsu = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

COLOR = (0,0,255)

contours, hiera = cv2.findContours(otsu, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

for cont in contours:
    x, y, w, h = cv2.boundingRect(cont)
    cv2.rectangle(temp, (x,y), (x+w, y+h), COLOR, 2)
# cv2.drawContours(temp, contour, -1, COLOR, 2)

cv2.imshow('img', img)
cv2.imshow('gray', gray)
cv2.imshow('otsu', otsu)
cv2.imshow('contour', temp)
cv2.waitKey(0)    
cv2.destroyAllWindows()
```
