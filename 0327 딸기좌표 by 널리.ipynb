```python
import cv2
import mediapipe as mp
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils
# For webcam input:
cap = cv2.VideoCapture('video3.mp4')
strawberry = cv2.imread('berry.jpg')
strawberry = cv2.resize(strawberry, (30, 30)) 

with mp_face_detection.FaceDetection(
    model_selection=0, min_detection_confidence=0.5) as face_detection:
  while cap.isOpened():
    success, image = cap.read()
    if not success:
      break

    h, w, _ = image.shape   # 영상 크기
    # To improve performance, optionally mark the image as not writeable to
    # pass by reference.
    image.flags.writeable = False
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = face_detection.process(image)
    # Draw the face detection annotations on the image.
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    if results.detections:
      for detection in results.detections:
        mp_drawing.draw_detection(image, detection)
        # 첫 번째 keypoint 가져오기
        for i in [0, 1]:
            keypoint = detection.location_data.relative_keypoints[i]
    
            # 비율 좌표 -> 실제 픽셀 좌표
            x = int(keypoint.x * w)
            y = int(keypoint.y * h)
    
            sh, sw, _ = strawberry.shape
    
             # 화면 밖으로 안 나가게 체크
            if y + sh < h and x + sw < w:
                image[y - sh//2 : y + sh//2, x - sw//2 : x + sw//2] = strawberry
    # Flip the image horizontally for a selfie-view display.
    cv2.imshow('MediaPipe Face Detection', image)
    if cv2.waitKey(60) & 0xFF == ord('q'):
      break
cap.release()
cv2.destroyAllWindows()
```
