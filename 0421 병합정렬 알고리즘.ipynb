```python
def merge_sort(arr):

    if len(arr) > 1:

        mid = len(arr) // 2          # 배열을 반으로 나눌 위치
        left_half = arr[:mid]        # 왼쪽 절반
        right_half = arr[mid:]       # 오른쪽 절반

        merge_sort(left_half)        # 왼쪽을 다시 정렬
        merge_sort(right_half)       # 오른쪽을 다시 정렬

        i = j = k = 0                # 위치 표시용 변수

        # 두 배열을 비교하면서 합치기
        while i < len(left_half) and j < len(right_half):

            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1

            else:
                arr[k] = right_half[j]
                j += 1

            k += 1

        # 왼쪽에 남은 값 처리
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        # 오른쪽에 남은 값 처리
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1


arr = [5, 3, 8, 4, 2, 7, 1, 6, 88, 3441, 12, 32454542, 64567656343]

merge_sort(arr)

print("정렬된 배열:", arr)
```
