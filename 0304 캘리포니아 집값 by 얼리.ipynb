```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
from sklearn.datasets import fetch_california_housing
```

- fetch_california_housing 함수를 이용해서 주택 데이터 불러오기

```python
data = fetch_california_housing()


```

- california 데이터 타입 확인

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df
```

- california 피처이름 확인

- 가져온 데이터를 데이터프레임으로 만들기

- california 데이터 확인

```python
df.shape
```

- 가져온 데이터를 판다스 데이터프레임으로 만들기

- 동네 중간 소득(MedInc)'과 '집값(Price)'의 상관계수 계산 ( corr함수 )

```python
df["Price"] = data.target   # 집값 컬럼 이름을 Price로 변경
```

- 산점도(Scatter) 그리기 
- x축은 소득(MedInc), y축은 집값(Price)
- alpha=0.1은 점을 투명하게 만들어서 겹치는 부분을 진하게 보여줌

```python
df = df.drop(columns=['corr'])
```

```python
df
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["MedInc"], df["Price"], alpha=0.1, color='red')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["AveBedrms"], df["Price"], alpha=0.1, color='purple')

plt.title("AveBedrms vs Price Scatter Plot (캘리포니아 주택 데이터)")
plt.xlabel("AveBedrms (침실의갯수)")
plt.ylabel("Price (집값)")
plt.grid(alpha=0.2)
plt.xlim(0,18)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["AveRooms"], df["Price"], alpha=0.1, color='green')

plt.title("MedInc vs Price Scatter Plot (캘리포니아 주택 데이터)")
plt.xlabel("AveRooms (방의갯수)")
plt.ylabel("Price (집값)")
plt.grid(alpha=0.2)
plt.xlim(0,50)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["Population"], df["Price"], alpha=0.1, color='green')

plt.title("Population vs Price Scatter Plot (캘리포니아 주택 데이터)")
plt.xlabel("Population (인구수)")
plt.ylabel("Price (집값)")
plt.grid(alpha=0.2)

plt.show()
```
